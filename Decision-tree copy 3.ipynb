```python
from sklearn import tree
import csv
```

```python
with open('TestingDataP2.csv', 'r') as file:

    stats_list = list(csv.reader(file))
    
labels = []
```

```python
for player in stats_list:
    print(player[1])
    if len(player[0]) > 0:
        labels.append("Yes")
    else:
        labels.append("No")
    player.pop(0)
    player.pop(0)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(stats_list,labels)
```

```python
print(clf.predict([[3,4,7,0.5,1,1,50]]))
```
